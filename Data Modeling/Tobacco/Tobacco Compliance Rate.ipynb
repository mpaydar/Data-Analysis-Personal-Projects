## Dataset:

**ECD Available Under Cap**:
        Number of new ECD licenses that can be issued in the Community District

## Multiple Regression Model

```python
from sklearn.linear_model import LinearRegression
first_model=LinearRegression()
X=df[['Electronic Cigarette Retail Dealer Cap','Active Electronic Cigarette Retail Dealer Licenses','Tobacco Compliance Rate','Tobacco Retail Dealer Cap' , 'Active Tobacco Retail Dealer Licenses']]
Y=df['Electronic Compliace Rate']
first_model.fit(X,Y)
optimum_y_intercept=first_model.intercept_
s1,s2,s3,s4,s5=first_model.coef_
Yhat=first_model.predict(X)
print("Optimum Y-Intercept for the model",optimum_y_intercept)
print("Optimum slope for the Electronic Cigarette Retail Dealer Cap",s1)
print("Optimum slope Active Electronic Cigarette Retail Dealer Licenses",s2)
print('Optimum slope Tobacco Compliance Rate',s3)
print('Optimum slope Tobacco Retail Dealer Cap',s4)
print('Optimum slope Active Tobacco Retail Dealer Licenses',s5)

```

## Assessment of the muli-linear regression model

```python
residuals=Y-Yhat
sns.residplot(x=Yhat, y=residuals, lowess=True, scatter_kws={'alpha': 0.5})
plt.xlabel('Predicted values (Yhat)')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

**From the residual plot, it is evidence that most of the prediction are close to the actual data point where we have most of our data points.As explored, majority of the is within 1.5 to 2.0 where most data lies.**

**Neglecting few exteme biased points on the plot, we see that the residual roughly forms a horizontal band indicating that the prediction of the data model is very close.**

# Assessment of the model

## R-Squared

```python
RSS=residuals**2
RSS=sum(RSS)
actual_mean=sum(Y)/len(Y)
TSS=sum((Y-actual_mean)**2)
R_Squared=1-(RSS/TSS)

print(R_Squared)
```

**The model captures about 79.69% of the total variability in the target variable, as measured by the sum of squares. This suggests that the model has a reasonably good fit to the data, explaining a substantial proportion of the observed variation. However for further analysis, I will do cross-validation to ensure the model prediction reliability as indicated by the R-Squared.**

## Cross-Validation

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error


k=5
kf = KFold(n_splits=k, shuffle=True, random_state=42)
# X = X.to_numpy()
mse_scores = []

for train_index, val_index in kf.split(X):
    X_train, X_val = X[train_index], X[val_index]
    y_train, y_val = y[train_index], y[val_index]
    
    # Fit the model on the training data
    first_model.fit(X_train, y_train)
    
    # Make predictions on the validation data
    y_pred = first_model.predict(X_val)
    
    # Calculate the mean squared error
    mse = mean_squared_error(y_val, y_pred)
    
    # Store the MSE score
    mse_scores.append(mse)

# Calculate the average MSE across all folds
average_mse = np.mean(mse_scores)

# Print the evaluation results
print("MSE scores:", mse_scores)
print("Average MSE:", average_mse)
```

**The MSE scores are all very close to zero, which suggests that the model's predictions are very accurate and closely aligned with the actual values. Averaging these small MSE values further confirms the consistency of the model's performance.**
